```python
import pyedflib
import numpy as np
from path import Path
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
```

```python
SLEEP_STAGES ={'Sleep stage ?'  :-1,
               'Sleep stage W'  :0,
               'Sleep stage N1' :1,
               'Sleep stage N2' :1,
               'Sleep stage N3' :2,
               'Sleep stage R'  :3,
               'Sleep stage W'  :0,
               'Sleep stage Light Sleep' :1,
               'Sleep stage Deep Sleep' :2,
               'Sleep stage REM'  :3,
               'Unscored - Bad Signal': -1,
               'Unscored': -1}

def read_header_pyedflib(fname):
    header = pyedflib.highlevel.read_edf(fname)[2]
    anno = header['annotations']
    stages_list , duration_list = [], []
    for i in anno:
        stages_list.append(SLEEP_STAGES.get(i[2]))
        duration_list.append(int(i[1]))
    return stages_list, duration_list

def read_labels_pyedflib(fname):
    header = pyedflib.highlevel.read_edf(fname)[2]
    anno = header['annotations']
    stages_list, duration_list = [], []
    for i in anno:
        stages_list.append(SLEEP_STAGES.get(i[2]))
        duration_list.append(int(i[1]))

    stage_epochs_list = []
    for idx, duration in enumerate(duration_list):
        for i in range(duration//30):
            stage_epochs_list.append(stages_list[idx])
    return np.array(stage_epochs_list)

def postprocess_rule1(stages_list, duration_list):
    remove_stages_list = []
    for idx, sleep_stage in enumerate(stages_list):
        if sleep_stage == 0 and idx != 0 and idx != (len(stages_list)-1):
            if duration_list[idx] <= 60 and (stages_list[idx-1] == 1 or stages_list[idx-1] == 2):
                duration_list[idx+1] += duration_list[idx]
                remove_stages_list.append(idx)
    stages_list = [stage for idx, stage in enumerate(stages_list) if idx not in remove_stages_list]
    duration_list = [duration for idx, duration in enumerate(duration_list) if idx not in remove_stages_list]
    return stages_list, duration_list

def postprocess_rule2(stages_list, duration_list):
    remove_stages_list = []
    for idx, sleep_stage in enumerate(stages_list):
        if sleep_stage == 0 and idx != 0:
            if duration_list[idx] > 90 and (stages_list[idx-1] == 3):
                duration_list[idx-1] += duration_list[idx]
                remove_stages_list.append(idx)
    stages_list = [stage for idx, stage in enumerate(stages_list) if idx not in remove_stages_list]
    duration_list = [duration for idx, duration in enumerate(duration_list) if idx not in remove_stages_list]
    return stages_list, duration_list
    
def postprocess_rule3(stages_list, duration_list):
    remove_stages_list = []
    for idx, sleep_stage in enumerate(stages_list):
        if sleep_stage == 1 and idx != 0:
            if duration_list[idx] <= 180 and (stages_list[idx-1] == 3):
                duration_list[idx-1] += duration_list[idx]
                remove_stages_list.append(idx)      
    stages_list = [stage for idx, stage in enumerate(stages_list) if idx not in remove_stages_list]
    duration_list = [duration for idx, duration in enumerate(duration_list) if idx not in remove_stages_list]
    return stages_list, duration_list

def postprocess_rule4(stages_list, duration_list):
    remove_stages_list = []
    for idx, sleep_stage in enumerate(stages_list):
        if idx == 0 or idx == len(stages_list)-1:
            continue
        if sleep_stage == -1: # Unknown stage
            if stages_list[idx-1] == 0:
                duration_list[idx-1] += duration_list[idx]
                remove_stages_list.append(idx)
            else:
                duration_list[idx+1] += duration_list[idx]
                remove_stages_list.append(idx)             
    stages_list = [stage for idx, stage in enumerate(stages_list) if idx not in remove_stages_list]
    duration_list = [duration for idx, duration in enumerate(duration_list) if idx not in remove_stages_list]
    
    return stages_list, duration_list

def remove_duplicate_stage(stages_list, duration_list):
    remove_duplicate_stages_list = []
    for idx in range(len(stages_list)-1):
        if (stages_list[idx] == stages_list[idx+1]):
            duration_list[idx+1] += duration_list[idx]
            remove_duplicate_stages_list.append(idx)
    stages_list = [stage for idx, stage in enumerate(stages_list) if idx not in remove_duplicate_stages_list]
    duration_list = [duration for idx, duration in enumerate(duration_list) if idx not in remove_duplicate_stages_list]
    
    stage_epochs_list = []
    for idx, duration in enumerate(duration_list):
        for i in range(duration//30):
            stage_epochs_list.append(stages_list[idx])
    return np.array(stage_epochs_list)
```

```python
# Load scores from experts and model
expert_scoring_fname = '/Users/admin/Documents/unscored_stage/EP2(7f6b)_PSG_050721_EE052/PSG_050721_EE052_HoangLisaAnh.edf'
ground_truth_scores = read_labels_pyedflib(expert_scoring_fname)
model_scoring_fname = '/Users/admin/Documents/unscored_stage/EP2(7f6b)_PSG_050721_EE052/EE052_EB_DEV7f6b_1620406213/EP2_PSG_EE052_EB_DEV7f6b_1620406213_V2_4_2.edf'
model_stages, model_duration = read_header_pyedflib(model_scoring_fname)

# model_stages, model_duration = postprocess_rule1(model_stages, model_duration)
# model_stages, model_duration = postprocess_rule2(model_stages, model_duration)
# model_stages, model_duration = postprocess_rule3(model_stages, model_duration)
model_stages, model_duration = postprocess_rule4(model_stages, model_duration)
# Remove duplicate stages and parse sleep stages to a list
model_stages_post_processing = remove_duplicate_stage(model_stages, model_duration)

# Remove padding epochs from start and end of study
model_scores = read_labels_pyedflib(model_scoring_fname)
scored_epochs = np.where(model_scores != -1)[0]
first_scored_epoch = scored_epochs[0]
last_scored_epoch = scored_epochs[-1]
model_scores = model_scores[first_scored_epoch:last_scored_epoch+1]
ground_truth_scores = ground_truth_scores[first_scored_epoch:last_scored_epoch+1]
model_stages_post_processing = model_stages_post_processing[first_scored_epoch:last_scored_epoch+1]
```

```python
subject_id = Path(model_scoring_fname).name.split('.edf')[0]
hypnogram_y_axis_stage_map = {-1:4, 0:3, 3:2, 1:1, 2:0}

fig, axes = plt.subplots(3, figsize=(24, 12))
plt.suptitle(subject_id, size=18)
plt.subplots_adjust(hspace=0.5)

epoch_timestamps = []
consensus_hypnogram_points = []
model_hypnogram_points = []
smoothed_model_hypnogram_points = []
for epoch_i in range(min(len(ground_truth_scores), len(model_scores))):
    epoch_timestamps += [int(30*epoch_i), int(30*(epoch_i+1))]
    consensus_hypnogram_points += [hypnogram_y_axis_stage_map[ground_truth_scores[epoch_i]], hypnogram_y_axis_stage_map[ground_truth_scores[epoch_i]]]
    model_hypnogram_points += [hypnogram_y_axis_stage_map[model_scores[epoch_i]], hypnogram_y_axis_stage_map[model_scores[epoch_i]]]
    smoothed_model_hypnogram_points += [hypnogram_y_axis_stage_map[model_stages_post_processing[epoch_i]], hypnogram_y_axis_stage_map[model_stages_post_processing[epoch_i]]]

axes[0].plot(epoch_timestamps, consensus_hypnogram_points, color='firebrick')
axes[0].set_title('Consensus Scoring', size=14)

axes[1].plot(epoch_timestamps, model_hypnogram_points)
unscored_count = len(np.where(model_scores == -1)[0])
valid_inds = set(np.where(ground_truth_scores != -1)[0]).intersection(set(np.where(model_scores != -1)[0]))
valid_inds = np.array(sorted(list(valid_inds)))
acc = accuracy_score(ground_truth_scores[valid_inds], model_scores[valid_inds])
axes[1].set_title('Raw Model Scoring ({:.4f} agreement, {} unscored epochs)'.format(acc, unscored_count), size=14)

axes[2].plot(epoch_timestamps, smoothed_model_hypnogram_points)
unscored_count = len(np.where(model_stages_post_processing == -1)[0])
valid_inds = set(np.where(ground_truth_scores != -1)[0]).intersection(set(np.where(model_stages_post_processing != -1)[0]))
valid_inds = np.array(sorted(list(valid_inds)))
acc = accuracy_score(ground_truth_scores[valid_inds], model_stages_post_processing[valid_inds])
axes[2].set_title('Model + HMM Assited Scoring ({:.4f} agreement, {} unscored epochs)'.format(acc, unscored_count), size=14)

for ax in axes:
    ax.set_yticks([0,1,2,3,4])
    ax.set_yticklabels(['DS','LS','R','W','Unk'])
    ax.set_ylim([-1,5])
    ax.set_xlim([0,epoch_timestamps[-1]])

plt.show()
```
